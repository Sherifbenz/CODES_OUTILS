### 1.2.2. Download all videos from a playlist 

```python
from pytube import Playlist
import os
p = Playlist("https://youtube.com/playlist?list=PLns_NvX_ArjW3-Z0-duZuBMOggod_lOr9&feature=shared")
for index, video in enumerate(p.videos):
    if index >= 0:
        numero = index + 1
        print(numero)
        vid=video.streams.get_by_resolution("720p")
        # Si aucun flux 720p n'est trouvé, revenir à la résolution la plus élevée disponible
        if vid is None:
            vid = video.streams.get_highest_resolution()
        # S'assurer qu'un flux valide est trouvé
        if vid is not None:
            # Obtenir le répertoire courant
            dossier_telechargement = os.getcwd()
            # Obtenir le nom du fichier de la vidéo
            nom_fichier = vid.default_filename         
            chemin_complet = os.path.join(dossier_telechargement, nom_fichier)
            nom_fichier_final = f"{numero:02d} {nom_fichier}"
            
            # Si le fichier existe déjà, renommer le fichier
            compteur = 1
            while os.path.exists(chemin_complet):
                nom_fichier = vid.default_filename 
                base, extension = os.path.splitext(nom_fichier)
                nom_fichier = f"{numero:02d} {base} ({compteur}){extension}"
                nom_fichier_final = nom_fichier
                chemin_complet = os.path.join(dossier_telechargement, nom_fichier)
                compteur += 1
            
            # Télécharger la vidéo
            vid.download(output_path=dossier_telechargement, filename=nom_fichier_final)
            print(f"Vidéo téléchargée sous le nom : {nom_fichier_final}")
        else:
            print("Aucun flux vidéo valide trouvé.")
```

### 2.3.1. Download all videos from a channel

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import time
from random import randint
import requests
from pytubefix import YouTube

# open browser
driver = webdriver.Firefox()

# youtube channel url
url = "http://www.youtube.com/c/Mentalit%C3%A9Ing%C3%A9nieur"

#get youtube video
def get_video_youtube(driver, url):
    driver.get(url+"/videos")
    time.sleep(3)
    
    # Faire défiler la page jusqu'à la fin
    while True:
        prev_ht = driver.execute_script("return document.documentElement.scrollHeight;")
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        time.sleep(2)
        ht = driver.execute_script("return document.documentElement.scrollHeight;")
        
        # Arrêter si nous ne pouvons pas faire défiler plus bas
        if ht == prev_ht:
            break
    
    # Récupérer les liens des vidéos
    links = driver.find_elements(By.XPATH, '//*[@id="video-title"]')
    
    # Enregistrer les liens dans un fichier
    with open('somefile.txt', 'w') as the_file:
        for link in links:
            print(link.get_attribute("title"))
            print(link.get_attribute("href"))
            yt = YouTube(link.get_attribute("href"))
            hd = yt.streams.get_highest_resolution()
            hd.download("youtube")
            the_file.writelines(link.get_attribute("href")+'\n')
            sleep(3)


get_video_youtube(driver,url)
```
