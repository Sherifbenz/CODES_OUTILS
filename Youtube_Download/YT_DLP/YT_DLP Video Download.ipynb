## 3.2. Playlist

### 3.2.1. Download all videos from a playlist without numbering

```python
import yt_dlp

def download_all_videos_from_playlist(playlist_url):
    # Options de configuration pour yt-dlp
    ydl_opts = {
        'format': 'best',  # Télécharger la meilleure qualité disponible
        'outtmpl': '%(uploader)s/%(title)s.%(ext)s',  # Organiser les vidéos dans un dossier basé sur le nom du créateur
        'merge_output_format': 'mp4',  # Fusionner en MP4 si nécessaire
        'ignoreerrors': True,  # Ignorer les vidéos avec des erreurs
        'writesubtitles': False,  # Télécharger les sous-titres si disponibles
        'writeautomaticsub': False,  # Télécharger les sous-titres générés automatiquement
        #'subtitleslangs': ['fr', 'en'],  # Télécharger les sous-titres en français et anglais
        'postprocessors': [{
            'key': 'FFmpegVideoConvertor',
            'preferedformat': 'mp4'  # Convertir en MP4
        }],
        'playlistreverse': True,  # Télécharger du plus ancien au plus récent
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        # Télécharger toutes les vidéos de la playlist
        ydl.download([playlist_url])

# Remplacez l'URL ci-dessous par celle de la playlist YouTube souhaitée
playlist_url = 'https://youtube.com/playlist?list=PLvX5VVNKso90SsSbTNbq5FOcXBn93khYo&feature=shared'

# Appeler la fonction pour télécharger les vidéos
download_all_videos_from_playlist(playlist_url)
```
